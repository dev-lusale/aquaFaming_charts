```python
import numpy as np
from mpmath import mp

mp.dps = 50  # high precision

# Coefficients (designed to cause cancellation)
a = 1.0
b = 1e8
c = 1.0

# High-precision reference
disc = mp.sqrt(b**2 - 4*a*c)
x1_exact = (-b + disc) / (2*a)
x2_exact = (-b - disc) / (2*a)

# Naive computation
disc_np = np.sqrt(b**2 - 4*a*c)
x1_naive = (-b + disc_np) / (2*a)
x2_naive = (-b - disc_np) / (2*a)

# Stable computation
x1_stable = (-2*c) / (b + disc_np)
x2_stable = (-b - disc_np) / (2*a)

# Relative errors
def rel_err(approx, exact):
    return abs(approx - exact) / abs(exact)

print("Quadratic Formula Errors:")
print("x1 naive  :", rel_err(x1_naive, x1_exact))
print("x1 stable :", rel_err(x1_stable, x1_exact))
print("x2 naive  :", rel_err(x2_naive, x2_exact))
print("x2 stable :", rel_err(x2_stable, x2_exact))
```
