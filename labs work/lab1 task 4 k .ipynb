```python
import numpy as np

def kahan_sum(arr):
    total = np.float32(0.0)
    c = np.float32(0.0)  # compensation

    for x in arr:
        y = x - c
        t = total + y
        c = (t - total) - y
        total = t

    return total
```

```python
import time
from decimal import Decimal, getcontext

getcontext().prec = 50

N = 1_000_000

# Data (float32 as required)
arr32 = np.array([1.0/k for k in range(1, N+1)], dtype=np.float32)

# High precision reference
exact = float(sum(Decimal(1)/Decimal(k) for k in range(1, N+1)))


# -------------------------
# Naive sum
# -------------------------
start = time.time()

naive = np.float32(0.0)
for x in arr32:
    naive += x

t_naive = time.time() - start


# -------------------------
# Kahan sum
# -------------------------
start = time.time()

kahan = kahan_sum(arr32)

t_kahan = time.time() - start


# -------------------------
# NumPy sum
# -------------------------
start = time.time()

numpy_sum = np.sum(arr32)

t_numpy = time.time() - start


# -------------------------
# Relative error
# -------------------------
def rel_err(val):
    return abs(val - exact) / exact


print("=== Accuracy ===")
print(f"Naive   error: {rel_err(naive):.3e}")
print(f"Kahan   error: {rel_err(kahan):.3e}")
print(f"NumPy   error: {rel_err(numpy_sum):.3e}")

print("\n=== Timing (seconds) ===")
print(f"Naive   time: {t_naive:.4f}")
print(f"Kahan   time: {t_kahan:.4f}")
print(f"NumPy   time: {t_numpy:.4f}")
```
